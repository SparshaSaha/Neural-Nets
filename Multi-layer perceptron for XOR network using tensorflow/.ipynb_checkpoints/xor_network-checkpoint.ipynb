```python
#Info on the Network

#This code uses tensor flow to plot a very popular problem of non-linear classification i.e, The Xor Problem
#By default i have used 2 nodes in input layer, 3 neurons in the hidden layer and 2 nodes in the output layer
#This program classifies the inputs into 2 classes namely 0 and 1
#The network can be modified in the In[1] under the tag #define the network

#The code also plots the decision boundary for the above problem

#Have fun!!! ;) 
```

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

#define the network
nodes_input=2
nodes_hidden_1=3
nodes_output=2
batch_size=1;

x=tf.placeholder('float',[None,2])
y=tf.placeholder('float')


#Define weights for neural network
hidden_l1={'weights':tf.Variable(tf.random_normal([nodes_input,nodes_hidden_1])),'biases':tf.Variable(tf.random_normal([nodes_hidden_1]))}
    
output_layer={'weights':tf.Variable(tf.random_normal([nodes_hidden_1,nodes_output])),'biases':tf.Variable(tf.random_normal([nodes_output]))}
```

```python
def get_data(xh):     #function to read data from file
    datafile=open(xh,"r")
    datalist=[]
    for line in datafile:
        z=line.split(",")
        xg=[float(i) for i in z]
        datalist.append(xg)
    f=np.matrix(datalist)
    return datalist
```

```python
def xor_nn(data):
    
    #feed forward throught the network
    l1=tf.add(tf.matmul(data,hidden_l1['weights']),hidden_l1['biases'])
    l1=tf.nn.relu(l1)
    
    output=tf.add(tf.matmul(l1,output_layer['weights']),output_layer['biases'])
    return (output)
```

```python
def train_neural_network(train_x,train_y):
    prediction=xor_nn(x)
    cost=tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits=prediction,labels=y))

    optimizer=tf.train.GradientDescentOptimizer(0.3).minimize(cost)

    hm_epochs=100

    with tf.Session() as sess:
        sess.run(tf.initialize_all_variables())
        saver=tf.train.Saver();
        

        for epoch in range(hm_epochs):
            epoch_loss=0
            for t in range(4):
                _,c=sess.run([optimizer,cost],feed_dict={x:[train_x[t]],y:[train_y[t]]})
                epoch_loss+=c
                
                correct=tf.equal(tf.argmax(prediction,1),tf.argmax(y,1))
                accuracy=tf.reduce_mean(tf.cast(correct,'float'))
            print('Epoch ',epoch,'conpleted ',hm_epochs,'loss: ',epoch_loss)
        
        save_path = saver.save(sess, "./mymodel")
        
        z=[]
        for i in xrange(-10,110):
            for j in xrange(-10,110):
                z.append([float(i/100.0),float(j/100.0)])
    
        k=z
        
    
        pre=sess.run(prediction,feed_dict={x:k})
        
        fin_list=[]
        for q in pre:
            zi=[]
            if(q[0]>q[1]):
                zi.append(1)
                zi.append(0)
            else:
                zi.append(0)
                zi.append(1)
            fin_list.append(zi)
            
        for i in range(len(k)):
            if(fin_list[i][0]==1):
                plt.plot(k[i][0],k[i][1],'r.')
        plt.show()
        
        
    return saver
    
    
        
        
```

```python
ox=get_data("x_data.txt")
oy=get_data("y_data.txt")

saver=train_neural_network(ox,oy)
```
